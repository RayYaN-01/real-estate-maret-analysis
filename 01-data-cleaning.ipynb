```python
import re
```

```python
# Function to convert single values into numeric format
def convert_to_number(value_str):

    value_str = value_str.strip()

    number = float(re.findall(r"[\d\.]+", value_str)[0])

    # Convert Crores
    if "Cr" in value_str:
        return number * 10000000

    # Convert Lakhs
    elif "L" in value_str:
        return number * 100000

    # Plain numeric values
    else:
        return number


# Main cleaning function
def clean_price(price):

    # Handle null values
    if pd.isnull(price):
        return np.nan

    price = str(price).strip()

    # Handle "Price on Request"
    if "Price on Request" in price:
        return np.nan

    try:

        # Handle ranges
        if "-" in price:

            parts = price.split("-")

            first_part = parts[0].strip()
            second_part = parts[1].strip()

            # If first part has no unit,
            # inherit unit from second part
            if ("Cr" not in first_part) and ("L" not in first_part):

                if "Cr" in second_part:
                    first_part += " Cr"

                elif "L" in second_part:
                    first_part += " L"

            first_value = convert_to_number(first_part)
            second_value = convert_to_number(second_part)

            # If first value is zero,
            # use second value directly
            if first_value == 0:
                return second_value

            # Otherwise take average
            return (first_value + second_value) / 2

        # Handle single values
        else:
            return convert_to_number(price)

    except:
        return np.nan
```

```python
# Apply cleaning function
realty_df['PRICE_CLEAN'] = realty_df['PRICE'].apply(clean_price)
```

```python
def clean_area(area):

    if pd.isnull(area):
        return np.nan

    area = str(area).replace("sq.ft.", "").strip()

    try:

        # Handle ranges
        if "-" in area:

            parts = area.split("-")

            first_value = float(parts[0].strip())
            second_value = float(parts[1].strip())

            return (first_value + second_value) / 2

        # Handle single values
        else:
            return float(area)

    except:
        return np.nan
```

```python
realty_df['AREA_CLEAN'] = realty_df['AREA'].apply(clean_area)
```
